# Open Banking data analytics

```python
def read_data(file):
    return pd.read_excel(file)

#Deriving Financial Analytics from Open Banking Dataset
def calculate_metrics(df):
    expenses = ['Rent', 'Food', 'Car', 'Water', 'Gas and Electric', 'Cellular', 'Council Tax', 'Entertainment', 'Other']
    df['Total Expenditure'] = df[expenses].sum(axis=1)
    df['Expense to Income Ratio'] = df['Total Expenditure'] / df['Income']
    df['Housing Expense Ratio'] =  df['Rent'] / df['Income']
    df['Debt to Income Ratio'] = (df['Total Expenditure'] - df['Rent']) / df['Income']
    df['Monthly Savings'] = df['Monthly Savings']
    df['Emergency Fund Months'] = df['Savings Balance'] / df['Total Expenditure']
    df['Year-over-Year Savings Growth'] = df['Savings Balance'].pct_change(periods=12) * 100 
    return df

#Generalized Qualitative Risk Assessment 
def assess_income_risk(df):
    conditions = [(df['Expense to Income Ratio'] < .5), (df['Expense to Income Ratio'] >= .5)]
    choices = ['Lower Risk', 'Higher Risk']
    df['Investment Profile'] = pd.cut(df['Expense to Income Ratio'], bins=[0, .5, 1], labels=choices)
    return df

#Quantitative Risk Assessment:
#Calculates a financial Well-Being Score using by assigning a quantitative 

'''Calculates a financial Well-Being Score by assigning a score to the value of each of the following:
   Emergency Fund, Expense to Income Ratio, Housing Expense Ratio, Debt to Income Ratio
   Emergency Fund rating accounts for 30% of the Score Rating,
   Housing Expense ratio accounts for 25% of the rating
   Debt to Income ratio accounts for 25% of the rating
   Total Expense to income ratio accounts for 20% of the rating
   
   Concept behind score assignment: The higher the rating for each category indicates 
   poorer financial well-being based on that specific metric indicating increased financial vulnerability
   
   High Rating = Bad Financial Position
   Low Rating = Strong Financial Position
'''
def calculate_financial_score(df):
    conditions_fund = [
        (df['Emergency Fund Months'] < 3),
        (df['Emergency Fund Months'] >= 3) & (df['Emergency Fund Months'] < 6),
        (df['Emergency Fund Months'] >= 6)
    ]
    scores_fund = [10, 7, 3]
    df['Emergency Fund: Score'] = np.select(conditions_fund, scores_fund, default=0)
    
    conditions_expense = [
        (df['Expense to Income Ratio'] < 0.28),
        (df['Expense to Income Ratio'] >= 0.28) & (df['Expense to Income Ratio'] < 0.5),
        (df['Expense to Income Ratio'] >= 0.5)
    ]
    scores_expense = [3, 7, 10]
    df['Expense to Income: Score'] = np.select(conditions_expense, scores_expense, default=0)

    conditions_housing = [
        (df['Housing Expense Ratio']  < .28),
        (df['Housing Expense Ratio']  >=.28) & (df['Housing Expense Ratio'] < .36),
        (df['Housing Expense Ratio']  >=.36) & (df['Housing Expense Ratio']  < .5),
        (df['Housing Expense Ratio']  >= .5)
    ]
    scores_housing = [2,4,6,8]
    df['Housing Expense Ratio: Score'] = np.select(conditions_housing, scores_housing, default=0)
    
    conditions_debt = [
        (df['Debt to Income Ratio'] < .35),
        (df['Debt to Income Ratio'] >= .35) & (df['Debt to Income Ratio'] < .43),
        (df['Debt to Income Ratio'] >= .43) & (df['Debt to Income Ratio'] < .5),
        (df['Debt to Income Ratio'] >= .5)
    ]
    scores_debt = [2,4,6,8]
    df['Debt Expense Ratio: Score'] = np.select(conditions_debt, scores_debt, default=0)
    #Print Statement to verify ratings
#     print(df['Emergency Fund Months'], df['Expense to Income Ratio'], df['Housing Expense Ratio'], df['Debt to Income Ratio'])
    
    #Calculating Financial Score 
    df['Financial Score'] = ((df['Emergency Fund: Score']*.3) + (df['Expense to Income: Score']*.20) + (df['Housing Expense Ratio: Score']*.25) + (df['Debt Expense Ratio: Score']*.25))
    return df

    #Calculating the Adjusted Risk Aversion Score:
    '''Questionnaire Score accounts for 65% of the score, Financial Rating accounts for 35% of the rating
    
        If users financial score is less than their questionnairre Score,
        Their risk aversion is not adjusted lower, and their risk aversion stays at the same level
        professed in the questionnaire.
        '''
def calculate_risk_aversion(df, questionnaire_score):
    df['Adjusted Risk Score'] = df['Financial Score'].apply(lambda x: questionnaire_score if x < questionnaire_score 
        else int(np.ceil(x * 0.35 + questionnaire_score * 0.65)) 
        if (x * 0.35 + questionnaire_score * 0.65) % 1 >= 0.5 
        else int(np.floor(x * 0.35 + questionnaire_score * 0.65)) #Rounding rules
    )
    df['Adjusted Risk Score'] = df['Adjusted Risk Score'].clip(lower=1, upper=10)
    return df

def main(file_path, questionnaire_score):
    data = read_data(file_path)
    income_metrics = calculate_metrics(data)  # Calculate metrics modifies 'data'
    risk_assessment = assess_income_risk(income_metrics)  # Pass modified data
    financial_scores = calculate_financial_score(risk_assessment)  # Continue passing modified data
    risk_scores = calculate_risk_aversion(financial_scores, questionnaire_score)  # Final modification
    
    # Storing results in a dictionary for easy access later
    results = {
        'income_metrics': income_metrics,
        'risk_assessment': risk_assessment,
        'financial_scores': financial_scores,
        'risk_scores': risk_scores
    }  
    return results

if __name__ == "__main__":
    file_path = 'Clients.xlsx'
    questionnaire_score = 1  # Example value
    results = main(file_path, questionnaire_score)
    # Ensuring future Accessibility
    print(results['income_metrics'][['Month', 'Expense to Income Ratio', 'Monthly Savings','Emergency Fund Months', 'Year-over-Year Savings Growth']])
    print(results['risk_scores'][['Adjusted Risk Score']])
```

# Step 1: 
## Calculating Standard Deviation for ETFs

Calculating standard Deviation is a financial industry standard for measuring the volatility of an asset.
Standard deviation is also useful because it can be leveraged to find further other metrics such as an assets variance and 
covariance, each of whom are more nuanced measures of volatility, that are more useful in some contexts.

$$
Standard Deviation = \sqrt{\frac{\sum_{i=1}^{N}(x_i - \bar{x})^2}{N-1}}
$$

Standard Deviation measures how much an assets performance differs from its average. 

In the equation above, we calculate standard deviation by
1. Finding the difference between each data point and its average
2. Squaring the Difference
3. Dividing by N-1, = # of data points
4. Taking the square root of the value from steps 1,2, and 3

To annualize our value we do an additional step:
5. Multiple the value from step 4 by the square root of the number of periods in the year.

Annualizing the standard deviation is a step taking to scale its value, making it more contextually meaningful across larger periods.

```python
#We multiply this by the square root of 250, because we assume that on average there are 250 trading days per year
Annualized_SD = ETF_df.pct_change().apply(lambda x: np.log(1+x)).std().apply(lambda x: x*np.sqrt(250))
Annualized_SD.plot(kind = 'bar')
```

# Step 2: 
## Calculating the Covariance:

###            Covariance is the measurment between the relationship between two types of variables

In modern portfolio theory the covariance of two assets is used to measure how two assets move together. 

    Negative Cov: Assets A and B move in opposite directions
    Zero Cov: The performance of the assets are unrelated
    Positive Cov: Assets A and B move in the same direction

The covariance matrix is a key metric used for risk management and allowing for diversification in the portfolio.

Assets with zero or negative correlation are more desirable, investors can  reduce their overall risk profile by offsetting periods of poor performance an asset may experience.

It is important to note that although selecting assets with zero or negative correlation, we do NOT eliminate risks associated with the market, nor do we eliminate the risk of each individual asset.

```python
ETF_df_returns = ETF_df.pct_change() # Calculated as: (val[1] - val[0])/val[0]
Cov_Matrix = ETF_df_returns.cov()
Cov_Matrix
```

# Step 4:
## Calculating Portfolio Volatility

Using a sample portfolio of equal weights, we demonstrate how we can calculate the volatility of the portfolio by leveraging its variance and standard deviation.

### Variance

Portfolio variance is the metric used in our program to measure volatility, defined as the square of standard deviation.

Variance measures how the performance of an asset varies around its mean.
    Higher Variation = Higher Volatility = Higher Risk

$$
\text{Portfolio Variance} = \sum_{i=1}^{n} \sum_{j=1}^{n} w_i w_j \text{Cov}(R_i, R_j)
$$

In the Equation above, we calculate portfolio variance by:
1. Multiplying the product of all possible combination of weights of assets in the set of sample portfolios, with the covariance of the assets in the sample portfolios.
2. Summing the products. 

Weights Wi and Wj, represent the proportion that asset i and asset j have within the portfolio. Wi and Wj tell us how much each assett contributes to the overall volatility of the portfolio. 

Cov(Ri, Rj) allows us to measure the relationship between assets i and j, thus multiplying their weights in the portfolio by their covariance, we can observe how changes in the performance of an asset influence the risk of the portfolio. 

```python
#Calculating portfolio variance
weights = np.ones(len(ETF_df_returns.columns))/len(ETF_df_returns.columns)
Sample_Portfolio_Var = Cov_Matrix.mul(weights, axis = 0).mul(weights, axis = 1).sum().sum()
Sample_Portfolio_Var
```

$$
\text{Portfolio Standard Deviation} = \sqrt{\text{Portfolio Variance}}
$$

Below is the Annual Standard Deviation for our sample Portfolio

```python
Sample_Portfolio_Annual_STD = np.sqrt(Sample_Portfolio_Var)
Sample_Portfolio_Annual_STD
```

# Step 5: 
## Calculate Expected Return

The expected return of an asset is used to measure the relative reward associated with taking on the risk of investing in an asset. 

$$
Expected Return = \sum_{i=1}^{n} w_i \times R_i
$$

In the equation above, we calculate expected retur by:
1. Multipling the weight of the each asset in the portfolio with its mean return
2. Sum the product from step 1

```python
#Because we are looking at annual returns we calculate the annual return at the end of each year.
df = ETF_df.resample('Y').last().pct_change()
#Gives us the average return for each asset
Expected_Return = df.mean()
#Dot is a np method which computes the sum of the product of 2 numpy arrays, 
#here we are multiplying the weights of each asset by their expected return
Sample_Portfolio_Return = np.dot(weights, Expected_Return)
Expected_Return
```

# Step 6: 
## Building the Efficiency Frontier

The efficient frontier is graphical representation of a subset of optimized portfolios consisting of only risky assets. The optimal portfolio for a given expected return is found by finding the we right combination of weights for each asset which minimize volatility. It represents the tradeoff between risks and return for sample portfolios and is crucial for identidying optimal portfolios.

In the code below we utilize the Markowitz Mean-Variance Optimization technique to build our efficiency frontier.

### Implementation Approach:
1. Identify a range of expected returns based off of the minimum and maximum expected return of assets in our basket.

2. Calculate the utility of each portfolio: Port_Utility() Function
    Utility is a quantitive value used to express value an investor gets from an investment
   
   The utility an investor derives from a portfolio is defined by the following equation:
$$ \text{Utility} = \text{Portfolio Return} - 0.5 \times \text{Risk Aversion} \times \text{Portfolio Variance} $$

### We measure the tradeoff between risk and return of two assets

Subtracting .5 x Risk Aversion x Portfolio Variance from the expected return of the portfolio serves as a risk adjustment factor ensuring that if a portfolio has higher volatility, the overall desirability of the portfolio is reduced, with respect to their professed risk aversion, aligning with the users goal of minimizing risk.
    
    
### Target_Return_Constraint():

The funcion enforces a constraint on the expected return value of the portfolio, ensuring that it aligns with the users investment goals

When called, this function can be used to ensure that the expected performance of an asset aligns with the users professed financial goals, however, in our code given that our optimization goal is to maximize utility, a specific target return is not required to build our efficiency frontier, and it is not used in our program. Nonetheless, if the program was changed to make the optimization goal achieving a specific return, then this function can be used.
    
Optimizing for utility rather than achieving a specific return creates room for the consideration of risk preferences. If an investor has an aggressive investment goal, achieving it may require taking on riskier investments, which may not align with their risk aversion. Therefore we optimize users utility to ensure the Roboadvisor always acts in accordance to users aversion. 

#### Future Works
In future works, if we identify users in strong financial positions who have high risk tolerance, then the optimization goal can be adjusted to achieving their specific investment goal, rather than maximizing utility.

### Constructing the Efficient Frontier:
Using the minimize function, we optimize the utility of each portfolio with the goal of minimizing utility, we accomplish this task using the Sequential Least Squares Programming method
##### SLSQP: 
Iterative optimization technique which finds the best solution for a minimization or maximization problem with respect for any restraints. In our problem, if we adjusted the optimization goal to achieving a target return, then the SLSQP cconstraint would be achieving the desired expected return.

Given we do not have a target return constraint in our progam, SLSQP simply changes the weights of the assets in the portfolio so that it achieves a given expected return while minimizing risk. 
##### KEY ASSUMPTION: GIVEN 2 PORTFOLIOS WITH THE SAME EXPECTED RETURN, USERS PREFER THE PORTFOLIO WITH THE LOWER VOLATILITY,  i.e. LOWER RISK

This assumption informs our optimization strategy and explains the hyperbolic shape of the efficiency frontier, as users would only choose to invest in portfolios along this frontier, as portfolios along this frontier minimize risk for each given expected return.

We then plot the volatility against their expected return for each portfolio to build the frontier

```python
# Mean Variance for building Efficiency Frontier
# We create a linearly spaced array of target returns, to ensure that the 
# target returns range from the lowest single asset expected return to the highest. 
# This way, we systematically cover the potential return spectrum that could be 
# achieved by various combinations of the assets.

def port_utility(weights, expected_returns, cov_matrix, risk_aversion):
    portfolio_return = np.dot(weights, expected_returns)
    portfolio_var = np.dot(weights.T, np.dot(cov_matrix, weights))
    utility = portfolio_return - .5 * risk_aversion * portfolio_var
    return -utility

def target_return_constraint(weights, expected_returns, target_return):
    # Constraint function for achieving a target return
    return np.dot(weights, expected_returns) - target_return

initial_weights = np.ones(len(Expected_Return)) / len(Expected_Return)
bounds = [(0, 1)] * len(initial_weights)  # Non-negative weights

# Generating potential returns between the minimum and maximum expected returns
potential_returns = np.linspace(min(Expected_Return), max(Expected_Return), 100)

Efficient_Port_Return = []
Efficient_Port_Vol = []
Port_Weights = []
risk_aversion = 5

for target_return in potential_returns:
    constraints = [{'type': 'eq', 'fun': lambda w: np.sum(w) - 1},  # weights sum to 1
                   {'type': 'eq', 'fun': lambda w: target_return_constraint(w, Expected_Return, target_return)}]  # achieve target return
    result = minimize(port_utility, initial_weights, args=(Expected_Return, Cov_Matrix, risk_aversion),
                      method='SLSQP', bounds=bounds, constraints=constraints)
    if result.success:
        #Save the successful identification of return, volatility, and weight
        Efficient_Port_Return.append(target_return)
        portfolio_var = np.dot(result.x.T, np.dot(Cov_Matrix, result.x))
        portfolio_vol = np.sqrt(portfolio_var) * 10
        Efficient_Port_Vol.append(portfolio_vol)
        Port_Weights.append(result.x)
        #x = weights
        #T = variance

# Plotting the Efficient Frontier
plt.scatter(Efficient_Port_Vol, Efficient_Port_Return, c='blue', label='Efficient Frontier')
plt.title('Efficient Frontier')
plt.xlabel('Portfolio Volatility (Standard Deviation)')
plt.ylabel('Expected Portfolio Return')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Observing the weights of each asset in the portfolios
data = {'Returns': Efficient_Port_Return, 'Vol': Efficient_Port_Vol}
data

for counter, symbol in enumerate(ETF_df.columns.tolist()):
    data[symbol+' weight'] = [w[counter] for w in Port_Weights]
    
portfolios = pd.DataFrame(data)
portfolios
```

### Optimal Risky Portfolio
The optimal risky portfolio is the portfolio which offers the optimal tradeoff of between risk and volatility. To discover this portfolio we find the sharpe ratio of each portfolio, and choose the portfolio with the highest Sharpe ratio.

### Sharpe Ratio
The Sharpe ratio helps us assess the performance of an asset against its risk, telling us how much additional return an investor gets for increased levels of risk. It effectively helps us identify the portfolio with the most justifiable level of risk given its expected return. Some risk can be justified if they are offset by the prospect of higher expected returns.

The sharpe ratio also compares the performance of a portfolio against a risk-free investment, such as government bonds, to help us compare our risk-taking invetsment activity against non risk taking investment strategies which also yields positive returns.

$$
\text{Sharpe Ratio} = \frac{\text{Portfolio Returns} - \text{Risk-Free Rate}}{\text{Portfolio Volatility}}
$$

In the equation above, we find the sharpe ratio of each asset by:
1. Finding the difference in the portfolio return and risk free rate
2. Dividing by portfolio volatility

Although the sharpe ratio does not have a finite scale, any score greater than 1 is deemed acceptable

The higher the sharpe ratio, the more desirable the portfolio
If the expected return of a portfolio is less than the risk free rate, then the Sharpe Ratio will be negative, indicating that there is no justification in takin on this risk and one would be better off investing their money in a risk free asset. 

```python
# Optimal Risky Portfolio - Sharpe Ratio

#Sharpe Ratio = \frac{{E(r_m) - r_f}}{{\sigma_m}}

risk_free = .035 
portfolios['Sharpe Ratio'] = ((portfolios['Returns'] - risk_free)/portfolios['Vol'])
orp = portfolios.iloc[portfolios['Sharpe Ratio'].idxmax()]
print(orp['Returns'])
print(orp['Vol'])

plt.scatter(portfolios['Vol'], portfolios['Returns'])
plt.scatter(mvp[1], mvp[0], color = 'red')
plt.scatter(orp[1], orp[0], color = 'yellow')
plt.xlabel('Portfolio Volatility')
plt.ylabel('Portfolio Return')
plt.title('Efficiency Frontier: MVP and ORP')
plt.grid(True)
plt.show()
```

# Step 8:
## Constructing the Capital Allocation Line

The Capital Allocation combines the optimal allocation of risky assets found on the Efficiency Frontier, with a risk free asset. This combination forms a linear array of portfolios that investors can invest on based on their risk aversion. 

#### The capital allocation line represents the optimal combination of risk-free and risky assets, users will invest along this line.

Each portfolio along the capital alloation line is defined by the following equation:

$$
\text{Capital Allocation Line:} \quad \text{rf} + \text{sharpe\_ratio} \times \text{sp}
$$

In the equation above we calculate the expected return of the portfolio by adding the expected return of a risk free asset, with the product of the sharpe ratio and standard deviation of the portfolio. By adding the risk free asset to our optimization problem, we are able to further assess the risk return tradeoff aiding in portfolio optimization and allowing for diversification through investment in risk-free assets.

```python
# # Deriving User Utility
# ## Utility Function
# #### A = Risk Aversion: The reward an investor requires to accept additional risk
# High A = Highly Risk Adverse
# Low A = Risk Loving
# #### Sigma = Volatility of the Asset
# $$ U = E(R) - 0.5A\sigma^2 $$
# ## Capital Allocation Line
# $$ E(r_p) = r_f + \frac{{E(r_m) - r_f}}{{\sigma_m}} \cdot \sigma_p \cdot A $$
#     rf = risk free rate  
#     sharpe ratio of optimal risky portfolio  
#     standard deviation of entire portfolio - we will solve for this value  

risk_free
risk_aversion

rm = orp['Returns'] # Return of Risky Portfolio
sm = orp['Vol'] # Volatility of Risky Portfolio
sharpe_ratio_orp = (rm - risk_free) / (sm)
weight_risky = 1 / user_income_risk_aversion
utility = []

def cal (rf, sharpe_ratio, sp):
    return rf + sharpe_ratio * sp

#Setting range of CAL to include 0
if not np.isclose(Efficient_Port_Vol[0], 0):
    Efficient_Port_Vol = np.insert(Efficient_Port_Vol, 0,0)

# sp = given level of portfolio volatility
# rf = riskfree weight
# Sharpe rato of the ORP

# Weight of Optimal Risky is a function of risk aversion and sharpe ratio 
# of the risky asset. Im using risk aversion 
# to determine how much should be allocated to the risky asset
# Higher A = Lower Weight_Risky -> more into risk free asset
# weight_risky = (rm-risk_free) / (risk_aversion*(sm**2))

#Calculating Capital Allocation Line Values for a given Volatility
#Incorporating the allocation of weight of the risky asset to determine
#expected return of each portfolio
cal_values = []
utility = []
for sp in Efficient_Port_Vol:
    cal_return = cal(risk_free, sharpe_ratio_orp, sp)
#     portfolio_return = risk_free + weight_risky * (sharpe_ratio_orp * sp)
    cal_values.append(cal_return)
    util = cal_return - (.5 *user_income_risk_aversion*(weight_risky*sp)**2)
    utility.append(util)

plt.scatter(portfolios['Vol'], portfolios['Returns'])
plt.scatter(mvp[1], mvp[0], color = 'red')
plt.scatter(orp[1], orp[0], color = 'yellow')
plt.plot(Efficient_Port_Vol,cal_values, label = 'Capital Allocation Line', color = 'orange')
plt.xlabel('Portfolio Volatility')
plt.ylabel('Portfolio Return')
plt.title('Efficiency Frontier')
plt.show()
```

# Portfolio Optimization:

In this part of the project we utilize the tools we gained through our journey in building out a top-down implementation of modern portfolio theory, to create optimal portfolios for our users

Key Ideas:
1. The weight of the risky asset class in a users portfolio is equal to 1 / Users Risk Aversion
    
    Why: This creeates a relationship in our program where users with higher risk aversion, invest less money in risky assets, than those who have high risk tolerance.

2. Risk Aversion Scale: 1 - 10
    
    1 = Strong Risk Tolerance   
    10 = Strong Risk Aversion  
    
    
3. If a user has perfectly strong risk aversion, i.e. Risk Aversion = 1, then their optimal portfolio is the Optimal Risky Portfolio
    
    Why: In this scenario the user has no risk aversion, thus they have no desire to invest in a risk free asset, therefore they will invest all of their money into risky assets, and the ORP is the portfolio along our efficiency frontier with the highest Sharpe Ratio and trade off of risk for reward.
    
In the cal_point() function we construct a modified version of our capital allocation line with 10 unique options for portfolios, but the one suggested to users is the portfolio which aligns with their risk aversion.

The variable Risk_Aversion_Levels generates an arrray of integers from 1 - 10 that represents a given risk aversion, we correspond these risk aversion values with specific positions along our capital allocation line.

The 10 positions along our capital allocation line represent the 10 potential portfolios a user can have be offered and demonstrates how risk preferences influences investment strategies.

```python
def cal_point(weight_risky):
    """Calculate a point on the CAL given the weight in the risky asset."""
    expected_return = risk_free + weight_risky * sharpe_ratio_orp * orp['Vol']
    volatility = weight_risky * orp['Vol']
    return expected_return, volatility

# Generating portfolios for different levels of risk aversion
risk_aversion_levels = np.arange(1, 11)  # From 1 to 10
portfolio_positions = [cal_point(1 / ra) for ra in risk_aversion_levels]

# Extracting returns and volatilities for plotting
returns = [pos[0] for pos in portfolio_positions]
volatilities = [pos[1] for pos in portfolio_positions]

plt.scatter(portfolios['Vol'], portfolios['Returns'], label='Efficiency Frontier', color = 'gray')
plt.scatter(mvp['Vol'], mvp['Returns'], color='red', label='Minimum Variance Portfolio')
plt.plot(volatilities, returns, label='Suggested Portfolios', color='blue', marker='o')
plt.xlabel('Portfolio Volatility')
plt.ylabel('Portfolio Return')
plt.title('Efficient Frontier and Suggested Portfolios')
plt.axhline(y=risk_free, color='green', linestyle='--', label='Risk-Free Rate')
plt.scatter(orp['Vol'], orp['Returns'], color='yellow', label='Optimal Risky Portfolio')
plt.legend()
plt.grid(True)
plt.show()
```

# Portfolio Optimization Continued:

Because we have calculated all key metrics and values in our implementation of modern portfolio theory, we can use them as inputs into our portfolio construction function 'find_optimal_portfolio'

find_optimal_portfolio() takes 5 inputs:
- 
    - portfolios: DataFrame containing all sample portfolios and their metrics
    - orp: Optimal Risky Portfolio
    - risk_free_rate: risk-free rate, .035%, UK Treasury bonds
    - risk_aversion: the user's income adjusted risk aversion score (1 to 10)
    - asset_names: assets in the portfolio
    
It delivers the optimal portfolio by:
First: Calculating the weight of risky and risk-free asset classes in the portfolio based on the users risk aversion
Second: Determining the expected return and volatility for the portfolio
Third: Calculates the weight of each asset in the portfolio
Fourth: Visualize the portfolio on our capital allocation line

```python
def find_optimal_portfolio(portfolios, orp, risk_free_rate, risk_aversion, asset_names):
    """    
    Parameters:
    - portfolios: DataFrame containing all sample portfolios and their metrics
    - orp: Optimal Risky Portfolio
    - risk_free_rate: risk-free rate, .035%, UK Treasury bonds
    - risk_aversion: the user's income adjusted risk aversion score (1 to 10)
    - asset_names: assets in the portfolio

    Returns:
    - Dataframe with of the optimal portfolio
    """
    
    #Decay Factor: Ensure a more even spread of portfolios along the Capital allocation line
    decay = -.5
    # Calculating the weight of the risky portfolio
    weight_risky = 1 / (risk_aversion + decay)
    weight_risk_free = 1 - weight_risky
    
    # Calculating the expected return and volatility
    expected_return = risk_free_rate + weight_risky * (orp['Returns'] - risk_free_rate)
    volatility = weight_risky * orp['Vol']
    
    # Calculating the weight of each asset in the overall portfolio
    asset_weights = {name: round(orp[name + ' weight'] * weight_risky, 2) for name in asset_names}
    
    # Creating dataframe for asset details
    asset_details = pd.DataFrame({
        'Asset': list(asset_weights.keys()),
        'Weight': list(asset_weights.values())
    })
    
    # Creating DataFrame for optimal portfolio
    portfolio_details = pd.DataFrame({
        'Risk-Free Asset Weight': weight_risk_free,
        'Risky Asset Total Weight': weight_risky,
        'Expected Return': expected_return,
        'Volatility': volatility
    }, index=[0])
    
    plt.figure(figsize=(10, 6))
    plt.scatter(portfolios['Vol'], portfolios['Returns'], color='gray', label='All Portfolios')
    plt.scatter(orp['Vol'], orp['Returns'], color='yellow', label='Optimal Risky Portfolio')
    plt.scatter(volatility, expected_return, color='red', marker='*', s=200, label='User Optimal Portfolio')
    plt.plot([0, orp['Vol']], [risk_free_rate, orp['Returns']], 'b--', label='Capital Allocation Line')
    plt.xlabel('Volatility')
    plt.ylabel('Expected Return')
    plt.title('Capital Allocation Line with User Optimal Portfolio')
    plt.axhline(y=risk_free_rate, color='green', linestyle='--', label='Risk-Free Rate')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return asset_details, portfolio_details

risk_free_rate = risk_free
risk_aversion = user_income_risk_aversion  # Example risk aversion
asset_names = ETF_list  # Example asset names

asset_details, optimal_portfolio_details = find_optimal_portfolio(portfolios, orp, risk_free_rate, risk_aversion, asset_names)

print("Asset Details:")
print(asset_details)
print("\nOptimal Portfolio Details:")
print(optimal_portfolio_details)
```
