# User data analytics

```python
def read_data(file):
    return pd.read_excel(file)

def calculate_metrics(df):
    expenses = ['Rent', 'Food', 'Car', 'Water', 'Gas and Electric', 'Cellular', 'Council Tax', 'Entertainment', 'Other']
    df['Total Expenditure'] = df[expenses].sum(axis=1)
    df['Expense to Income Ratio'] = df['Total Expenditure'] / df['Income']
    df['Housing Expense Ratio'] =  df['Rent'] / df['Income']
    df['Debt to Income Ratio'] = (df['Total Expenditure'] - df['Rent']) / df['Income']
    df['Monthly Savings'] = df['Monthly Savings']
    df['Emergency Fund Months'] = df['Savings Balance'] / df['Total Expenditure']
    df['Year-over-Year Savings Growth'] = df['Savings Balance'].pct_change(periods=12) * 100 
    return df

def assess_income_risk(df):
    conditions = [(df['Expense to Income Ratio'] < .5), (df['Expense to Income Ratio'] >= .5)]
    choices = ['Lower Risk', 'Higher Risk']
    df['Investment Profile'] = pd.cut(df['Expense to Income Ratio'], bins=[0, .5, 1], labels=choices)
    return df

def calculate_financial_score(df):
    conditions_fund = [
        (df['Emergency Fund Months'] < 3),
        (df['Emergency Fund Months'] >= 3) & (df['Emergency Fund Months'] < 6),
        (df['Emergency Fund Months'] >= 6)
    ]
    scores_fund = [10, 7, 3]
    df['Emergency Fund: Score'] = np.select(conditions_fund, scores_fund, default=0)
    
    conditions_expense = [
        (df['Expense to Income Ratio'] < 0.28),
        (df['Expense to Income Ratio'] >= 0.28) & (df['Expense to Income Ratio'] < 0.5),
        (df['Expense to Income Ratio'] >= 0.5)
    ]
    scores_expense = [3, 7, 10]
    df['Expense to Income: Score'] = np.select(conditions_expense, scores_expense, default=0)

    conditions_housing = [
        (df['Housing Expense Ratio']  < .28),
        (df['Housing Expense Ratio']  >=.28) & (df['Housing Expense Ratio'] < .36),
        (df['Housing Expense Ratio']  >=.36) & (df['Housing Expense Ratio']  < .5),
        (df['Housing Expense Ratio']  >= .5)
    ]
    scores_housing = [2,4,6,8]
    df['Housing Expense Ratio: Score'] = np.select(conditions_housing, scores_housing, default=0)
    
    conditions_debt = [
        (df['Debt to Income Ratio'] < .35),
        (df['Debt to Income Ratio'] >= .35) & (df['Debt to Income Ratio'] < .43),
        (df['Debt to Income Ratio'] >= .43) & (df['Debt to Income Ratio'] < .5),
        (df['Debt to Income Ratio'] >= .5)
    ]
    scores_debt = [2,4,6,8]
    df['Debt Expense Ratio: Score'] = np.select(conditions_debt, scores_debt, default=0)

    df['Financial Score'] = ((df['Emergency Fund: Score'] + df['Expense to Income: Score'] + df['Housing Expense Ratio: Score'] +  df['Debt Expense Ratio: Score']) / 4)
    return df

def calculate_risk_aversion(df, questionnaire_score):
    df['Adjusted Risk Score'] = df['Financial Score'].apply(lambda x: questionnaire_score if x < questionnaire_score 
        else int(np.ceil(x * 0.35 + questionnaire_score * 0.65)) 
        if (x * 0.35 + questionnaire_score * 0.65) % 1 >= 0.5 
        else int(np.floor(x * 0.35 + questionnaire_score * 0.65)) #Rounding rules
    )
    df['Adjusted Risk Score'] = df['Adjusted Risk Score'].clip(lower=1, upper=10)
    return df

def main(file_path, questionnaire_score):
    data = read_data(file_path)
    income_metrics = calculate_metrics(data)  # Calculate metrics modifies 'data'
    risk_assessment = assess_income_risk(income_metrics)  # Pass modified data
    financial_scores = calculate_financial_score(risk_assessment)  # Continue passing modified data
    risk_scores = calculate_risk_aversion(financial_scores, questionnaire_score)  # Final modification
    
    # Storing results in a dictionary for easy access later
    results = {
        'income_metrics': income_metrics,
        'risk_assessment': risk_assessment,
        'financial_scores': financial_scores,
        'risk_scores': risk_scores
    }  
    return results

if __name__ == "__main__":
    file_path = 'Clients.xlsx'
    questionnaire_score = 6  # Example value
    results = main(file_path, questionnaire_score)
    # Example of how to access a specific variable later in the code
    print(results['income_metrics'][['Month', 'Expense to Income Ratio', 'Monthly Savings','Emergency Fund Months', 'Year-over-Year Savings Growth']])
    print(results['risk_scores'][['Adjusted Risk Score']])
```

# Calculating Standard Deviation for ETFs:

We utilize Standard Deviation, to measure the  volaility of ETFs
$$
Standard Deviation = \sqrt{\frac{\sum_{i=1}^{N}(x_i - \bar{x})^2}{N-1}}
$$


```python
# Utilizing Standard Deviation (Square Root of Variance) to measure the volatility of each ETF
#We multiply this by the square root of 250, because we assume that on average there are 250 trading days per year
Annualized_SD = ETF_df.pct_change().apply(lambda x: np.log(1+x)).std().apply(lambda x: x*np.sqrt(250))
Annualized_SD.plot(kind = 'bar')
```

# Calculating the Covariance:

###            Covariance is the measurment between the relationship between two types of variables
    Negative Cov: When Asset A is performing well, Asset B tends to perform poorly, and vice versa
    Zero Cov: The performance of the assets are unrelated
    Positive Cov: When Asset A is perform well, Asset B tends to perform well, and vice versa

```python
ETF_df_returns = ETF_df.pct_change() # Calculated as: (val[1] - val[0])/val[0]
Cov_Matrix = ETF_df_returns.cov()
Cov_Matrix
```

# Understanding the Covariance Matrix
### Diagonal
The diagonal yields the variance of each individual ETF
Outside of the diagonal lies the covariance of the ETF's with one another

## Why:
The covariance informs us on the the direction of the relationship between two assets. This is important because in MPT, we are tasked with developing a highly diversified portfolio. Negative covariance implies that the performances of assets have an inverse relaitonship. Adding negatively related assets to a portfolio reduce it's volatility/risk and improve its stability. 

# Calculating the Correlation Coefficient:

```python
Corr_Matrix = ETF_df_returns.corr()
```

## Why:
The correlation coefficient not only tells us the direction of the relationship between assets, but it also details details the strength of the relationship. Investors prefer to correlation coefficient because its output is standardized on a scale of -1 to 1.
### Interpretation:
    -1: Perfectly negative relationship
     0: Unrelated relationship
     1: Perfectly positive relationship

# Calculating Portfolio Variance and Standard Deviation

We generate a sample portfolio where each asset has an equal weight within the portfolio. We then calculate the volatility of the portfolio utilizing the variance and standard deviation formulas below. We utilize the covariance and correlaiton matrix for a sample portfolio

## Variance
$$
\text{Portfolio Variance} = \sum_{i=1}^{n} \sum_{j=1}^{n} w_i w_j \text{Cov}(R_i, R_j)
$$



```python
weights = np.ones(len(ETF_df_returns.columns))/len(ETF_df_returns.columns)
Sample_Portfolio_Var = Cov_Matrix.mul(weights, axis = 0).mul(weights, axis = 1).sum().sum()
Sample_Portfolio_Var
```

$$
\text{Portfolio Standard Deviation} = \sqrt{\text{Portfolio Variance}}
$$

Below is the Annual Standard Deviation for our sample Portfolio

```python
Sample_Portfolio_Annual_STD = np.sqrt(Sample_Portfolio_Var)
Sample_Portfolio_Annual_STD
```

# Calculating Investment Portfolio Expected Return

$$
Expected Return = \sum_{i=1}^{n} w_i \times R_i
$$

Sum of the weighted averages of the expect returns of each asset in our sample portfolio

```python
#Because we are looking at annual returns we calculate the annual return at the end of each year.
df = ETF_df.resample('Y').last().pct_change()
#Gives us the average return for each asset
Expected_Return = df.mean()
#Dot is a np method which computes the sum of the product of 2 numpy arrays, 
#here we are multiplying the weights of each asset by their expected return
Sample_Portfolio_Return = np.dot(weights, Expected_Return)
Expected_Return
```

```python
# def portfolio_volatility(weights,cov_matrix):
#     Portfolio_Var = cov_matrix.mul(weights, axis = 0).mul(weights, axis = 1).sum().sum()
#     Port_SD = np.sqrt(Portfolio_Var)*np.sqrt(250)
#     return Port_SD

# def random_weights(etfs):
#     weights = np.random.random(len(ETF_df.columns))
#     return weights / np.sum(weights)

# def random_portfolio(expected_returns, cov_matrix):
#     etfs = len(expected_returns)
#     weights = random_weights(etfs)
#     rand_vol = portfolio_volatility(weights, cov_matrix)
#     rand_return = np.dot(weights, expected_returns)
#     return weights, rand_return, rand_vol

# def target_return_max(weights, expected_returns, target_return):
#     return np.dot(weights, expected_returns) - target_return

# Port_Return = []
# Port_Vol = []
# Port_Weights = []
# tot_ports = 100

# for i in range(tot_ports):
#     weights, rand_return, rand_vol = random_portfolio(Expected_Return, Cov_Matrix)
#     Port_Return.append(rand_return)
#     Port_Vol.append(rand_vol)
#     Port_Weights.append(weights)

# # Calculate the efficiency frontier
# Efficient_Port_Return = []
# Efficient_Port_Vol = []

# potential_returns = np.linspace(min(Expected_Return), max(Expected_Return), 100)

# #Ensuring non negative 
# initial_weights = np.ones(len(Expected_Return)) / len(Expected_Return)
# bounds = [(0,1)] * len(initial_weights)

# for pot_return in potential_returns:
#     constraints = ({'type': 'eq','fun': target_return_max, 'args': (Expected_Return, pot_return)},
#                    {'type': 'eq', 'fun': lambda w: np.sum(w) - 1})
#     result = minimize(portfolio_volatility, initial_weights, args=(Cov_Matrix,),method ='SLSQP', bounds=bounds, constraints=constraints)
#     if result.success:
#         Efficient_Port_Return.append(pot_return)
#         Efficient_Port_Vol.append(result.fun)
        
# # Plot the results
# plt.scatter(Efficient_Port_Vol, Efficient_Port_Return, label='Efficiency Frontier')
# plt.xlabel('Portfolio Volatility')
# plt.ylabel('Portfolio Return')
# plt.title('Efficiency Frontier')
# plt.legend()
# plt.show()
```

```python
# Mean Variance for building Efficicny Frontier
# Create a linearly spaced array of target returns, ensuring that the 
# target returns range from the lowest single asset expected return to the highest. 
# This way, we systematically cover the potential return spectrum that could be 
# achieved by various combinations of the assets.

# def portfolio_volatility(weights, cov_matrix):
#     # Calculating portfolio variance
#     Portfolio_Var = np.dot(weights.T, np.dot(cov_matrix, weights))
#     # Convert variance to standard deviation and annualize it
#     Port_SD = np.sqrt(Portfolio_Var) * np.sqrt(250)
#     return Port_SD

def port_utility(weights, expected_returns, cov_matrix, risk_aversion):
    portfolio_return = np.dot(weights, expected_returns)
    portfolio_var = np.dot(weights.T, np.dot(cov_matrix, weights))
    utility = portfolio_return - .5 * risk_aversion * portfolio_var
    return -utility

def target_return_constraint(weights, expected_returns, target_return):
    # Constraint function for achieving a target return
    return np.dot(weights, expected_returns) - target_return

initial_weights = np.ones(len(Expected_Return)) / len(Expected_Return)
bounds = [(0, 1)] * len(initial_weights)  # Non-negative weights


# Generate potential returns between the minimum and maximum expected returns
potential_returns = np.linspace(min(Expected_Return), max(Expected_Return), 100)

Efficient_Port_Return = []
Efficient_Port_Vol = []
Port_Weights = []
risk_aversion = 5

for target_return in potential_returns:
    constraints = [{'type': 'eq', 'fun': lambda w: np.sum(w) - 1},  # weights sum to 1
                   {'type': 'eq', 'fun': lambda w: target_return_constraint(w, Expected_Return, target_return)}]  # achieve target return
    result = minimize(port_utility, initial_weights, args=(Expected_Return, Cov_Matrix, risk_aversion),
                      method='SLSQP', bounds=bounds, constraints=constraints)
    if result.success:
        #Save the successful identification of return, volatility, and weight
        Efficient_Port_Return.append(target_return)
        portfolio_var = np.dot(result.x.T, np.dot(Cov_Matrix, result.x))
        portfolio_vol = np.sqrt(portfolio_var) * 10
        Efficient_Port_Vol.append(portfolio_vol)
        Port_Weights.append(result.x)
        #x = weights
        #T = variance

# Plotting the Efficient Frontier
plt.scatter(Efficient_Port_Vol, Efficient_Port_Return, c='blue', label='Efficient Frontier')
plt.title('Efficient Frontier')
plt.xlabel('Portfolio Volatility (Standard Deviation)')
plt.ylabel('Expected Portfolio Return')
plt.legend()
plt.grid(True)
plt.show()
```

```python
data = {'Returns': Efficient_Port_Return, 'Vol': Efficient_Port_Vol}
data

for counter, symbol in enumerate(ETF_df.columns.tolist()):
    data[symbol+' weight'] = [w[counter] for w in Port_Weights]
    
portfolios = pd.DataFrame(data)
portfolios
```

```python
# Deriving the Minimum Variance Portfolio 

mvp = portfolios.iloc[portfolios['Vol'].idxmin()]
mvp
plt.scatter(portfolios['Vol'], portfolios['Returns'])
plt.scatter(mvp[1], mvp[0], color = 'red')
plt.xlabel('Portfolio Volatility')
plt.ylabel('Portfolio Return')
plt.title('Efficiency Frontier')
plt.show()

# Optimal Risky Portfolio - Sharpe Ratio

#Sharpe Ratio = \frac{{E(r_m) - r_f}}{{\sigma_m}}

risk_free = .035
portfolios['Sharpe Ratio'] = ((portfolios['Returns'] - risk_free)/portfolios['Vol'])
orp = portfolios.iloc[portfolios['Sharpe Ratio'].idxmax()]
print(orp['Returns'])
print(orp['Vol'])

plt.scatter(portfolios['Vol'], portfolios['Returns'])
plt.scatter(mvp[1], mvp[0], color = 'red')
plt.scatter(orp[1], orp[0], color = 'yellow')
plt.xlabel('Portfolio Volatility')
plt.ylabel('Portfolio Return')
plt.title('Efficiency Frontier')
plt.grid(True)
plt.show()
```

```python
# # Deriving User Utility
# ## Utility Function
# #### A = Risk Aversion: The reward an investor requires to accept additional risk
# High A = Highly Risk Adverse
# Low A = Risk Loving
# #### Sigma = Volatility of the Asset
# $$ U = E(R) - 0.5A\sigma^2 $$
# ## Capital Allocation Line
# $$ E(r_p) = r_f + \frac{{E(r_m) - r_f}}{{\sigma_m}} \cdot \sigma_p \cdot A $$
#     rf = risk free rate  
#     sharpe ratio of optimal risky portfolio  
#     standard deviation of entire portfolio - we will solve for this value  

risk_free
risk_aversion

rm = orp['Returns'] # Return of Risky Portfolio
sm = orp['Vol'] # Volatility of Risky Portfolio
sharpe_ratio_orp = (rm - risk_free) / (sm)
weight_risky = 1 / risk_aversion
utility = []

def cal (rf, sharpe_ratio, sp):
    return rf + sharpe_ratio * sp

#Setting range of CAL to include 0
if not np.isclose(Efficient_Port_Vol[0], 0):
    Efficient_Port_Vol = np.insert(Efficient_Port_Vol, 0,0)

# sp = given level of portfolio volatility
# rf = riskfree weight
# Sharpe rato of the ORP


# Weight of Optimal Risky is a function of risk aversion and sharpe ratio 
# of the risky asset. Im using risk aversion 
# to determine how much should be allocated to the risky asset
# Higher A = Lower Weight_Risky -> more into risk free asset
# weight_risky = (rm-risk_free) / (risk_aversion*(sm**2))

#Calculating Capital Allocation Line Values for a given Volatility
#Incorporating the allocation of weight of the risky asset to determine
#expected return of each portfolio
cal_values = []
utility = []
for sp in Efficient_Port_Vol:
    cal_return = cal(risk_free, sharpe_ratio_orp, sp)
#     portfolio_return = risk_free + weight_risky * (sharpe_ratio_orp * sp)
    cal_values.append(cal_return)
    util = cal_return - (.5 *risk_aversion*(weight_risky*sp)**2)
    utility.append(util)

plt.scatter(portfolios['Vol'], portfolios['Returns'])
plt.scatter(mvp[1], mvp[0], color = 'red')
plt.scatter(orp[1], orp[0], color = 'yellow')
plt.plot(Efficient_Port_Vol,cal_values, label = 'Capital Allocation Line', color = 'orange')
plt.xlabel('Portfolio Volatility')
plt.ylabel('Portfolio Return')
plt.title('Efficiency Frontier')
plt.show()
```

```python
# Optimal Portfolio = Largest Utility

#Identifying index with largest utility
max_util = np.argmax(utility)
util_vol = Efficient_Port_Vol[max_util]
util_ret = cal_values[max_util]

plt.scatter(portfolios['Vol'], portfolios['Returns'])
plt.scatter(mvp[1], mvp[0], color = 'red')
plt.scatter(orp[1], orp[0], color = 'yellow')
plt.plot(Efficient_Port_Vol,cal_values, label = 'Capital Allocation Line', color = 'orange')
plt.scatter(util_vol, util_ret, color='green', label='Optimal Portfolio')
plt.xlabel('Portfolio Volatility')
plt.ylabel('Portfolio Return')
plt.title('Efficiency Frontier')
plt.show()
```

```python
def cal_point(weight_risky):
    """Calculate a point on the CAL given the weight in the risky asset."""
    expected_return = risk_free + weight_risky * sharpe_ratio_orp * orp['Vol']
    volatility = weight_risky * orp['Vol']
    return expected_return, volatility

# Generate portfolios for different levels of risk aversion
risk_aversion_levels = np.arange(1, 11)  # From 1 to 10
portfolio_positions = [cal_point(1 / ra) for ra in risk_aversion_levels]

# Extract returns and volatilities for plotting
returns = [pos[0] for pos in portfolio_positions]
volatilities = [pos[1] for pos in portfolio_positions]

plt.scatter(portfolios['Vol'], portfolios['Returns'], label='Efficiency Frontier', color = 'gray')
plt.scatter(mvp['Vol'], mvp['Returns'], color='red', label='Minimum Variance Portfolio')
plt.plot(volatilities, returns, label='Suggested Portfolios', color='blue', marker='o')
plt.xlabel('Portfolio Volatility')
plt.ylabel('Portfolio Return')
plt.title('Efficient Frontier and Suggested Portfolios')
plt.axhline(y=risk_free, color='green', linestyle='--', label='Risk-Free Rate')
plt.scatter(orp['Vol'], orp['Returns'], color='yellow', label='Optimal Risky Portfolio')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def find_optimal_portfolio(portfolios, orp, risk_free_rate, risk_aversion, asset_names):
    """
    Given the user's risk aversion, find and visualize the optimal portfolio on the CAL.
    
    Parameters:
    - portfolios: DataFrame containing all portfolios and their metrics
    - orp: Series/DataFrame row with data for the Optimal Risky Portfolio
    - risk_free_rate: float, the risk-free rate
    - risk_aversion: int, the user's risk aversion score (1 to 10)
    - asset_names: list of strings, names of the assets in the portfolio

    Returns:
    - dict containing details of the optimal portfolio
    """
    
    #Add decay factor
    # Calculate the weight in the risky portfolio
    weight_risky = 1 / risk_aversion
    weight_risk_free = 1 - weight_risky
    
    # Calculate the expected return and volatility for this weight
    expected_return = risk_free_rate + weight_risky * (orp['Returns'] - risk_free_rate)
    volatility = weight_risky * orp['Vol']
    
    # Calculate the weight of each asset in the overall portfolio
    asset_weights = {name: orp[name + ' weight'] * weight_risky for name in asset_names}
    
    # Create a detailed dictionary of the portfolio
    portfolio_details = {
        'Risk-Free Asset Weight': weight_risk_free,
        'Risky Asset Total Weight': weight_risky,
        'Asset Weights': asset_weights,
        'Expected Return': expected_return,
        'Volatility': volatility
    }
    
    # Visualize the CAL and highlight the selected portfolio
    plt.figure(figsize=(10, 6))
    plt.scatter(portfolios['Vol'], portfolios['Returns'], color='gray', label='All Portfolios')
    plt.scatter(orp['Vol'], orp['Returns'], color='yellow', label='Optimal Risky Portfolio')
    plt.scatter(volatility, expected_return, color='red', marker='*', s=200, label='User Optimal Portfolio')
    plt.plot([0, orp['Vol']], [risk_free_rate, orp['Returns']], 'b--', label='Capital Allocation Line')
    plt.xlabel('Volatility')
    plt.ylabel('Expected Return')
    plt.title('Capital Allocation Line with User Optimal Portfolio')
    plt.axhline(y=risk_free_rate, color='green', linestyle='--', label='Risk-Free Rate')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return portfolio_details

risk_free_rate = risk_free
risk_aversion = user_income_risk_aversion  # Example risk aversion
asset_names = ETF_list  # Example asset names

optimal_portfolio_details = find_optimal_portfolio(portfolios, orp, risk_free_rate, risk_aversion, asset_names)
print(optimal_portfolio_details)
```
